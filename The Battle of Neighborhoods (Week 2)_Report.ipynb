## 3. Methodology 

The methodology used herein includes:
- finding top venues in each neighborhood;
- clustering the neighborhoods;
- creating a map based on such clusters; and
- examing ach cluster.

For the clustering of the neighborhoods, “K-Means Clustering Algorithm" has been applied. K-Means Clustering Algorithmis a type of unsupervised learning, which is used when you have unlabeled data (i.e., data without defined categories or groups). The goal of this algorithm is to find groups in the data, with the number of groups represented by the variable K. The algorithm works iteratively to assign each data point to one of K groups based on the features that are provided. Data points are clustered based on feature similarity. The results of the K-means clustering algorithm are the centroids of the K clusters, which can be used to label new data labels for the training data (each data point is assigned to a single cluster).

More detailed steps includes:
- Collectiing dataset including location and type (or category) of each restaurant within the designated latitude and longitude;
- Exploring the restaurant density across the different geographic areas by using K-Mean; and
- Finding the most promising areas within the created clusters of locations with less resturants in radius of 500 meters of the city center.
